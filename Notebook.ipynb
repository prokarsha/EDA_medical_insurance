```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'medical_insurance.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Summary Statistics

```python
data.describe()
```

### Distribution of categorical variables

```python
sns.countplot(data['sex'])
plt.title('Distribution of Gender')
plt.show()
```

```python
sns.countplot(data['smoker'])
plt.title('Distribution of Smokers')
plt.show()
```

```python
sns.countplot(data['region'])
plt.title('Distribution of Regions')
plt.show()
```

### Distribution of numerical variables

```python
sns.histplot(data['age'], kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.histplot(data['bmi'], kde=True)
plt.title('Distribution of BMI')
plt.show()
```

```python
sns.histplot(data['children'], kde=False)
plt.title('Distribution of Number of Children')
plt.show()
```

### Relationship between numerical variables and charges

```python
sns.scatterplot(x='age', y='charges', data=data)
plt.title('Age vs Charges')
plt.show()
```

```python
sns.scatterplot(x='bmi', y='charges', data=data)
plt.title('BMI vs Charges')
plt.show()
```

### Relationship between categorical variables and charges

```python
sns.boxplot(x='sex', y='charges', data=data)
plt.title('Gender vs Charges')
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=data)
plt.title('Smoker vs Charges')
plt.show()
```

```python
sns.boxplot(x='region', y='charges', data=data)
plt.title('Region vs Charges')
plt.show()
```
